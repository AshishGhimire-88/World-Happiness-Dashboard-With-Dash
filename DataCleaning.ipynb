Fixing the column name and filtering out the dataset for the year 2013 to 2022

```python
FactorReport.rename(columns={"Country Name": "Country"}, inplace=True)
FactorReport=FactorReport[FactorReport['Year'].isin(range(2013, 2023))]
HappinessIndex=HappinessIndex[HappinessIndex['Year'].isin(range(2013, 2023))]
```

```python
df=HappinessIndex.merge(FactorReport, how='inner', on=['Country', 'Year'])
```

```python
df.drop(columns=['Confidence In National Government'], inplace=True)
```

```python
df.to_csv("datasets/Dataset.csv")
```
